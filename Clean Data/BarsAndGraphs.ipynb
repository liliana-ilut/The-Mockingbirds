```python
#Import Dependencies 

import matplotlib
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import requests
from scipy.stats import linregress
```

```python
# Take in all of our data and read it into pandas

csvpath = "./Resources/AllMerge.csv"

combined_df = pd.read_csv(csvpath)
combined_df.head()
```

# 1st question: Can the amount of restaurants in a city be a predictor of a city's economic standing?

```python
# the x and y values (x=restaurant, y= income)
rest = combined_df ["Total CBP and NES Establishments"]
inc = combined_df["Median"]

#create the scatter plt, modify colors, grid and create labels for axis and title
plt.scatter(rest, inc,color="turquoise", edgecolors="black", alpha=0.75)
plt.xlabel("Restaurants")
plt.ylabel("Median Income (Dollars)")
plt.title("Restaurants vs Median Income")
plt.grid()

#save figure as .png file in Figures folder and show the figure on the script
plt.tight_layout()
plt.savefig("Figures/Restaurant_vs_Income.png")
plt.show()
```

```python
# the x and y values (x=restaurant, y= income)
rest = combined_df ["Total CBP and NES Establishments"]
inc = combined_df["Median"]

#create the linear regression equation
(slope, intercept, rvalue, pvalue, stderr) = linregress(rest, inc)
regress_values = rest * slope + intercept
line_eq = "y = " + str(round(slope,4)) + "x + " + str(round(intercept,4))
plt.scatter(rest, inc,color="turquoise", edgecolors="black", alpha=0.75)
plt.plot(rest,regress_values,"r-")
plt.annotate(line_eq,(10,50), fontsize = 15, color = "red")

#create the scatter plt, modify colors, grid and create labels for axis and title
plt.xlabel("Restaurant")
plt.ylabel("Median Income (Dollars)")
plt.title("Linear Regression Restaurant vs Median Income")
plt.grid()
print(f" Linear Equation is : {line_eq}")
print(f" R-squared is {rvalue**2}")
#save figure as .png file in Figures folder and show the figure on the script
plt.tight_layout()
plt.savefig("Figures/Regression_Restaurants_vs_Income.png")
plt.show()
```

```python
#create data frame wiht the top 5 and low 5 based on income
counties = ["WebbCounty,Texas", "HidalgoCounty,Texas", "CameronCounty,Texas", "BronxCounty,NewYork", "MarionCounty,Florida",
            "ArlingtonCounty,Virginia","SantaClaraCounty,California", "FairfaxCounty,Virginia", "HowardCounty,Maryland", "LoudounCounty,Virginia"]
median_income = [35659, 36176, 37061, 37525, 39383,  
                 110388, 111069, 115717, 120941, 134464 ]
CBPandNES_establishments = [591.0, 1483, 842, 2822, 470,
                         554.0, 3943, 1973 , 503.0, 647.0]
obesity_percentage= [37.7, 34.4, 30.9, 28.8, 30.7, 
                    20,19.2, 21.3,24.8, 22.1 ]
population= [275910, 865939, 423908, 1432132, 359977,
      237521,1937570, 1150795, 323196, 406850 ]
total_df = pd.DataFrame({ "Median Income": median_income, "Restaurants":CBPandNES_establishments, "Obesity Percentage": obesity_percentage, "Population": population}, index= counties)
total_df
```

```python
# create bar cahrt comparing the top 5 and the low 5 counties based on income and population
import numpy as np
import matplotlib.pyplot as plt
 
bars1 = total_df["Median Income"]
bars2 = total_df["Population"]

barWidth = 0.5
r1 = np.arange(len(bars1))
r2 = [x + barWidth for x in r1]

plt.bar(r1,bars1, color = "blue",width=barWidth, label="Median")
plt.bar(r2,bars2, color = "red", width=barWidth, label="Population")

plt.xlabel('Counties', fontweight='bold')
plt.xticks( [r + barWidth for r in range(len(bars1))],counties, rotation = 45)
plt.title("Populations and Median Income")

plt.legend()
plt.tight_layout()
plt.savefig("Figures/Bars_Population_Income.png")
plt.show()
```

```python
#create var chart comparing the same top 5 and low 5 counties as in the previous bar chart based on restaurants and obesity
bars1 = total_df["Restaurants"]
bars2 = total_df["Obesity Percentage"]


barWidth = 0.5

r1 = np.arange(len(bars1))
r2 = [x + barWidth for x in r1]

plt.bar(r1,bars1, color = "green",width=barWidth, label="Restaurants")
plt.bar(r2,bars2, color = "orange", width=barWidth,label="Obesity")

plt.xlabel('Counties', fontweight='bold')
plt.xticks( [r + barWidth for r in range(len(bars1))],counties, rotation = 45)

plt.tight_layout()
plt.legend()
plt.title("Restaurants and Obesity Percentage")
plt.tight_layout()
plt.savefig("Figures/Bars_Restaurants_Obesity.png")
plt.show()
```
