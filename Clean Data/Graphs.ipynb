```python
#Import Dependencies 

import matplotlib
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import requests
import time
from scipy.stats import linregress
import datetime as dt 
```

```python
# Take in all of our data and read it into pandas

csvpath = "./Resources/AllMerge.csv"

combined_df = pd.read_csv(csvpath)
combined_df.head(3)
```

```python
# the x and y values (x=population, y= obesity)
pop = combined_df ["Population"]
ob = combined_df["Obesity Percentage"]

#create the scatter plt, modify colors, grid and create labels for axis and title
plt.scatter(pop, ob,color="coral", edgecolors="grey", alpha=0.75)
plt.xlabel("Population")
plt.ylabel("Obesity Percentage")
plt.title("Population vs Obesity")
plt.grid()

#save figure as .png file in Figures folder and show the figure on the script
# plt.savefig("Figures/Plot_Lat_vs_Temp.png")
plt.show()
```

```python
# the x and y values (x=population, y= obesity)
pop = combined_df ["Population"]
ob = combined_df["Obesity Percentage"]

#create the linear regression equation
(slope, intercept, rvalue, pvalue, stderr) = linregress(pop, ob)
regress_values = pop * slope + intercept
line_eq = "y = " + str(round(slope,4)) + "x + " + str(round(intercept,4))

plt.scatter(pop, ob,color="coral", edgecolors="grey", alpha=0.75)
plt.plot(pop,regress_values,"r-")
plt.annotate(line_eq,(1,10), fontsize = 15, color = "red")

#create the scatter plt, modify colors, grid and create labels for axis and title

plt.xlabel("Population")
plt.ylabel("Obesity Percentage")
plt.title(" Linear Regression Population vs Obesity")
plt.grid()

print(f" Linear Equation is : {line_eq}")
print(f" R-squared is {rvalue**2}")

#save figure as .png file in Figures folder and show the figure on the script
# plt.savefig("Figures/Plot_Lat_vs_Temp.png")
plt.show()
```
